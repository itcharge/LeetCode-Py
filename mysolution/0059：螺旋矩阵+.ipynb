```python
from typing import List
import numpy as np

class Solution:
    def generateMatrix(self, n: int) -> List[List[int]]:
        rows = cols = n
        all = int(rows/2 + 0.5)              # 
        start = -1   #  00 -11 -22.
        res = [[1 for _ in range(cols)] for _ in range(rows) ]   # 
        val = 1  #

        for k in range(all):
            start += 1
            i = j = start    # 
            while ( j < cols - k):  # 
                res[i][j] = val
                val += 1
                j += 1
                
            j -= 1
            i += 1

            while(  i < rows - k): #   j = cols - k j - 1  len(res) 
                res[i][j] = val
                val += 1
                i += 1 
            i -= 1            
            j -=1 

            while( j >= k and rows!= 0):#    [[2,3]]
                res[i][j] = val
                val += 1
                j -= 1
            j += 1
            i -=1

            while(i >= k+1 and cols != 1):#
                res[i][j] = val
                val += 1
                i -=1
        return res

a = Solution()
a.generateMatrix(10)
```
