```python
#       0
# mask  0
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        m = len(matrix)
        n = len(matrix[0])
        
        # 0
        lines = [0] * m
        rows = [0] * n
        
        # 0
        for i in range(m):
            for j in range(n):
                if matrix[i][j] == 0:
                    lines[i] = 1  # i0
                    rows[j] = 1  # j0
        
        # 0
        for i in range(m):
            for j in range(n):
                if lines[i] or rows[j]:
                    matrix[i][j] = 0
```
