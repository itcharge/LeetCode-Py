```python

from typing import List


class Solution:
    def crackPassword(self, password: List[int]) -> str:
        #  
        # a + b < b + aa,b
        password = [str(num) for num in password]
        for m in range(len(password)-1):
            for j in range(len(password) - 1 -m):
                if (password[j]) + (password[j+1]) > (password[j+1]) + (password[j]):
                            password[j], password[j+1] = password[j+1], password[j]
        #  10
        
       # res = ''.join(map(str,password))  # ''
        return password

list1 = [15,9,6]
a = Solution()
a.crackPassword(list1)
```
